```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("bank_transactions.csv")
```

```python
df.head()
```

```python

import matplotlib.pyplot as plt

df['CustomerDOB'] = pd.to_datetime(df['CustomerDOB'])
df = df.drop(df[df['CustomerDOB'] == '1800-01-01'].index)
df.loc[df['CustomerDOB'].dt.year >= 2022, 'CustomerDOB'] = df['CustomerDOB'] - pd.DateOffset(years=100)
today = pd.to_datetime('today')
def calculate_age(born):
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
df['CustomerAge'] = df['CustomerDOB'].apply(calculate_age)
df.head()
```

```python
sns.histplot(df['CustomerAge'], bins=30, kde=True, color='lightgreen')
```

```python
location_counts = df['CustLocation'].value_counts().nlargest(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=location_counts.index, y=location_counts.values, palette='viridis')
plt.title('Top 10 Locations with the Highest Number of Customers')
plt.xlabel('Location')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
top_10_locations = df['CustLocation'].value_counts().nlargest(10)

# Visualize customer distribution across top 10 locations
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_locations.index, y=top_10_locations.values, palette='coolwarm')
plt.title('Customer Distribution Across Top 10 Locations')
plt.xlabel('Location')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```

```python
location_transaction_volumes = df.groupby('CustLocation')['TransactionAmount (INR)'].sum().nlargest(10).sort_values(ascending=False)

green_palette = sns.color_palette("Greens", len(location_transaction_volumes))

plt.figure(figsize=(10, 4))
sns.barplot(x=location_transaction_volumes.index, y=location_transaction_volumes.values, palette=green_palette)
plt.title('Transaction Volumes for Different Locations')
plt.xlabel('Location')
plt.ylabel('Total Transaction Amount (INR)')
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn import preprocessing
```

```python
prepro=preprocessing.LabelEncoder()
```

```python
df['CustGender']= prepro.fit_transform(df['CustGender'])
df['CustLocation']= prepro.fit_transform(df['CustLocation'])
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=['TransactionID', 'CustomerID', 'CustomerDOB','TransactionDate',
       'TransactionTime'],axis=0)
```

```python
X=df
```

```python
k = 3 
kmeans = KMeans(n_clusters=k)
kmeans.fit(X)
```

```python
cluster_labels = kmeans.labels_
```

```python
import numpy as np
unique_labels, counts = np.unique(cluster_labels, return_counts=True)
for label, count in zip(unique_labels, counts):
    print("Cluster {}: {} data points".format(label, count))
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['CustLocation'], df['TransactionAmount (INR)'], c=cluster_labels)
plt.xlabel("CustLocation")
plt.ylabel("TransactionAmount (INR)")
plt.title("K-Means Clustering")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['CustomerAge'], df['TransactionAmount (INR)'], c=cluster_labels)
plt.xlabel("CustomerAge")
plt.ylabel("TransactionAmount (INR)")
plt.title("K-Means Clustering")
plt.show()
```
